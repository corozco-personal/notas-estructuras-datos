```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de n
n_values = np.linspace(1, 30, 500)

# Funciones
def C1(n):
    return 2**(n+3) - 6

def C2(n):
    return (2**(n+3) - 6) + 3 * (2**(n+1) - 2)

# Calcular valores de las funciones
C1_values = C1(n_values)
C2_values = C2(n_values)

# Crear la gráfica
fig = plt.figure(figsize=(10, 4))
plt.plot(n_values, C1_values, label=r'$C(n) = 2^{n+3} - 6$', color='blue')
plt.plot(n_values, C2_values, label=r'$C(n) = 2^{n+3} - 6) + 3(2^{n+1} - 2)$', color='red')

# Etiquetas y título
plt.xlabel('n')
plt.ylabel('C(n)')
plt.title('Comparación de complejidad (serie de Fibonacci)')
plt.legend(loc='upper left')
plt.grid(True)

# Mostrar la gráfica
plt.show()

fig.tight_layout(pad=0)

# Guardar la figura con 300 DPI
fig.savefig('comparacion_complejidad_exponencial.png', dpi=300)
```
