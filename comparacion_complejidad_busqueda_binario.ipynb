```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de n
n_values = np.linspace(1, 100, 500)

# Funciones
def C1(n):
    return 3*(2 + 3*np.log2(n))

def C2(n):
    return 3*(2 + 3*np.log2(n)) + 1

# Calcular valores de las funciones
C1_values = C1(n_values)
C2_values = C2(n_values)

# Crear la gráfica
fig = plt.figure(figsize=(10, 4))
plt.plot(n_values, C1_values, label=r'$C(n) = 3[2 + 3\log_2(n)]$', color='blue')
plt.plot(n_values, C2_values, label=r'$C(n) = 3[2 + 3\log_2(n)] + 1$', color='red')

# Etiquetas y título
plt.xlabel('n')
plt.ylabel('C(n)')
plt.title('Complejidad Computacional de Búsqueda Binaria')
plt.legend(loc='lower right')
plt.grid(True)

# Mostrar la gráfica
plt.show()

fig.tight_layout(pad=0)

# Guardar la figura con 300 DPI
fig.savefig('comparacion_complejidad_binario_1.png', dpi=300)
```
