```python
import numpy as np
import matplotlib.pyplot as plt

# Valores de n
n_values = np.linspace(1, 1e10, 500)

# Funciones
def C1(n):
    return 5*n + 4

def C2(n):
    return 4*n + 6

# Calcular valores de las funciones
C1_values = C1(n_values)
C2_values = C2(n_values)

# Crear la gráfica
fig = plt.figure(figsize=(10, 4))
plt.plot(n_values, C1_values, label=r'$C(n) = 5n + 4$', color='blue')
plt.plot(n_values, C2_values, label=r'$C(n) = 4n + 6$', color='red')

# Etiquetas y título
plt.xlabel('n')
plt.ylabel('C(n)')
plt.title('Complejidad computacional de ciclo for simple')
plt.legend(loc='lower right')
plt.grid(True)

# Mostrar la gráfica
plt.show()

fig.tight_layout(pad=0)

# Guardar la figura con 300 DPI
fig.savefig('comparacion_complejidad_for.png', dpi=300)
```
