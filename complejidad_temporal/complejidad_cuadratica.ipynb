```python
#@title Utils

# FunciÃ³n para ordenar por burbuja (bubble sort)
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
```

```python
# Lista de tamaÃ±os de lista para probar
tamanos_lista = np.arange(1, 1001, 20)

# Inicializar arrays para los tiempos de bubble sort
tiempos_bubble_sort = np.zeros(len(tamanos_lista))

average = 20

# Medir el tiempo de bubble sort 5 veces para cada tamaÃ±o de lista
for i, n in enumerate(tamanos_lista):
    tiempos_bubble_sort_n = np.zeros(average)
    for j in range(average):
        lista = np.random.randint(0, 1000, size=n)
        start_time = time.time()
        bubble_sort(lista)
        end_time = time.time()
        tiempos_bubble_sort_n[j] = end_time - start_time
    tiempos_bubble_sort[i] = np.mean(tiempos_bubble_sort_n)

# Generar el cÃ³digo QR
qr_data = "https://bit.ly/3wF7Qsk"
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)
qr.add_data(qr_data)
qr.make(fit=True)
qr_img = qr.make_image(fill_color='black', back_color='white')

# Guardar la imagen del cÃ³digo QR
qr_img_path = "codigo_qr.png"
qr_img.save(qr_img_path)

# Cargar la imagen del cÃ³digo QR con PIL
qr_img_pil = Image.open(qr_img_path)

# Convertir la imagen de PIL a un formato que matplotlib puede manejar
qr_img_np = np.array(qr_img_pil)

# Graficar los resultados
fig, ax1 = plt.subplots(1, 1, figsize=(10, 5))

# GrÃ¡fica: Tiempo de bubble sort de elementos en un arreglo
ax1.plot(tamanos_lista, tiempos_bubble_sort, label='Tiempo de ejecuciÃ³n')

# Agregar la lÃ­nea T(n) = O(n^2)
n_squared = tamanos_lista**2
# Escalar n_squared para ajustarse a la escala de tiempos_bubble_sort
n_squared_scaled = n_squared * (tiempos_bubble_sort[-1] / n_squared[-1])
ax1.plot(tamanos_lista, n_squared_scaled, label='Complejidad teÃ³rica $T(n)=n^2$', linestyle='dotted', color='red')

ax1.set_xlabel('TamaÃ±o del arreglo ($n$)')
ax1.set_ylabel('Tiempo de ejecuciÃ³n [segundos]')
ax1.set_title('Complejidad temporal para algoritmo de ordenamiento bubble sort')
ax1.legend(loc='lower right')
ax1.grid(True)

# Formatear los ejes x e y en notaciÃ³n cientÃ­fica
ax1.xaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.yaxis.set_major_formatter(plt.ScalarFormatter(useMathText=True))
ax1.ticklabel_format(axis='x', style='sci', scilimits=(0, 0))
ax1.ticklabel_format(axis='y', style='sci', scilimits=(0, 0))

# Insertar la imagen del cÃ³digo QR en la grÃ¡fica
qr_position = [0.885, 0.23, 0.1, 0.25]
ax_inset = fig.add_axes(qr_position)
ax_inset.imshow(qr_img_np, cmap='gray')
ax_inset.axis('off')

plt.tight_layout()
plt.show()

fig.savefig('complejidad_temporal_cuadratico.png', dpi=300)
```
