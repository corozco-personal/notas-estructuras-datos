```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones f(n) y g(n)
def f(n):
    return n**3 + 2*n**2 + n + 5

def g(n):
    return n**3

# Constante c
c = 9

# Rango de valores de n
n = np.linspace(1, 2, 1000)

# Calcular los valores de las funciones
f_values = f(n)
g_values = c * g(n)

# Encontrar el valor aproximado de n0 donde f(n) <= c * g(n)
tolerance = 0.1  # Definir una tolerancia para la aproximacin
n0 = next((n[i] for i, (f_val, g_val) in enumerate(zip(f_values, g_values)) if np.isclose(f_val, g_val, atol=tolerance)), None)

# Crear la grfica
fig, ax = plt.subplots(figsize=(10, 4))

# Configurar la etiqueta de g(n) segn el valor de c
if c==1:
  label_text = '$c \\cdot g(n) = n^3$'
  title_text = f'Comparacin entre $f(n)$ y $g(n)$ (Big-O)'
else:
  label_text = f'$c \\cdot g(n) = {c}n^3$'
  title_text = f'Comparacin entre $f(n)$ y ${c}g(n)$ (Big-O)'

# Graficar las funciones
ax.plot(n, g_values, label=label_text, color='red')
ax.plot(n, f_values, label='$f(n) = n^3 + 2n^2 + n + 5$', color='blue')

# Marcar el valor de n0 si existe
if n0:
    ax.axvline(x=n0, color='green', linestyle='--', label=f'$n_0 \\approx {round(n0, 2)}$')
    ax.fill_between(n, f_values, g_values, where=(f_values <= g_values), interpolate=True, color='red', alpha=0.3)

# Etiquetas y ttulo
ax.set_xlabel('$n$')
ax.set_ylabel('Valor de la funcin')
ax.set_title(title_text)
ax.legend(loc='upper left')
ax.grid(True)

# Ajustar diseo para eliminar espacios en blanco innecesarios
fig.tight_layout(pad=0)

# Mostrar la grfica
plt.show()

# Guardar la grfica como archivo PNG
fig.savefig('comparacion_big_O.png', dpi=300)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones f(n) y g(n)
def f(n):
    return n**3 + 2*n**2 + n + 5

def g(n):
    return n**3

# Constantes c1 y c2
c1 = 1
c2 = 9

# Rango de valores de n
n = np.linspace(1, 2, 1000)

# Calcular los valores de las funciones
f_values = f(n)
g_values_lower = c1 * g(n)
g_values_upper = c2 * g(n)

# Crear la grfica
fig, ax = plt.subplots(figsize=(10, 4))

# Configurar las etiquetas de g(n) segn los valores de c1 y c2
label_text_lower = f'$c_1 \\cdot g(n) = {c1}n^3$'
label_text_upper = f'$c_2 \\cdot g(n) = {c2}n^3$'
title_text = 'Comparacin entre $f(n)$ y $g(n)$ (Big-$\\Theta$)'

# Graficar las funciones
ax.plot(n, g_values_lower, label=label_text_lower, color='green')
ax.plot(n, g_values_upper, label=label_text_upper, color='red')
ax.plot(n, f_values, label='$f(n) = n^3 + 2n^2 + n + 5$', color='blue')

# Marcar el valor de n0 si existe
n0 = 1
ax.axvline(x=n0, color='purple', linestyle='--', label=f'$n_0 = {n0}$')
ax.fill_between(n, g_values_lower, g_values_upper, where=(g_values_lower <= f_values) & (f_values <= g_values_upper), interpolate=True, color='blue', alpha=0.2)

# Etiquetas y ttulo
ax.set_xlabel('$n$')
ax.set_ylabel('Valor de la funcin')
ax.set_title(title_text)
ax.legend(loc='upper left')
ax.grid(True)

# Ajustar diseo para eliminar espacios en blanco innecesarios
fig.tight_layout(pad=0)

# Guardar la grfica como archivo PNG
fig.savefig('comparacion_big_theta.png', dpi=300, bbox_inches='tight')
plt.show()
```
